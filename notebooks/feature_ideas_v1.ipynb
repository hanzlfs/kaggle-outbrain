```python
#### for example, after normalization 
def normalize_freq(x):
    lst = x[1].split('|')
    return '|'.join([y.split(':')[0] + ':' + str(int(float(y.split(':')[1])/float(x[0]) * 100)) for y in lst])
### for example, the column 'page_views:document_id:categ_topI-freq'
df[['page_views:uuid:num_visits','page_views:document_id:categ_topI-freq']].apply(lambda x: normalize_freq(x), axis = 1).head()
```

```python
#### Split events:geo_location into country>state>DMA,create columns events:geo:country, events:geo:state, events:geo:DMA
def split_geo(x):
    if x is None or x == '-99':
        return ['-99'] * 3
    lst = x.split('>')
    if lst:
        country = lst[0]
    state = lst[1] if len(lst) > 1 else '-99'
    DMA = lst[2] if len(lst) > 2 else '-99'
    return [country, state, DMA]        
    
df.loc[:,['events:geo_location']].applymap(lambda x: split_geo(x))
#df.loc[:, ['events:geo:country','events:geo:state', 'events:geo:DMA']] =\
#    df.loc[:,['events:geo_location']].applymap(lambda x: split_geo(x), axis = 1)

```

```python
#### Split page_views:geo_location into country:freq > state:freq > DMA:freq,\
#### create columns page_views:geo:country_freq, page_views:geo:state_freq, page_views:geo:DMA_freq
def split_geo_pageview(x):
    segs = x.split('|') # segmentations

    dic = {}
    dic['country'] = {}
    dic['state'] = {}
    dic['DMA'] = {}
    
    for seg in segs:        
        freq = int(seg.split(':')[1])
        country, state, DMA = split_geo(seg.split(':')[0]) 
        if country not in dic['country']:
            dic['country'][country] = 0
        if state not in dic['state']:
            dic['state'][state] = 0
        if DMA not in dic['DMA']:
            dic['DMA'][DMA] = 0
        dic['country'][country] += freq
        dic['state'][state] += freq
        dic['DMA'][DMA] += freq
    return ['|'.join(sorted([k + ':' + str(v) for k, v in dic[key].items()])) for key in ['country','state','DMA']]
          

df.loc[:,['page_views:uuid:geo_location-freq']].applymap(lambda x: split_geo_pageview(x))
```
